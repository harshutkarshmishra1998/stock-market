1 - Import Packages

```python
import numpy as np
import pandas as pd
```

2 - Load Dataset

```python
df = pd.read_csv("stock-indexes.csv")
df.head()
```

3 - Preprocessing

```python
# Converting the stock indexes to returns since we SARIMAX model needs stationary data for proper training
df_ret = pd.DataFrame([])
df_ret['date'] = df.Date
df_ret['bse'] = df.bse.pct_change(1).mul(100)
df_ret['spx'] = df.spx.pct_change(1).mul(100)
df_ret['ftse'] = df.ftse.pct_change(1).mul(100)
df_ret['dax'] = df.dax.pct_change(1).mul(100)
df_ret['nikkei'] = df.nikkei.pct_change(1).mul(100)
df_ret.head()
```

```python
df_ret = df_ret.iloc[1:]
df_ret.head()
```

```python
# To renove the upper layer of zeroes
df_ret = df_ret.iloc[1:]
df_ret.head()
```

```python
df_ret = df_ret.reset_index(drop=True)
df_ret.head()
```
